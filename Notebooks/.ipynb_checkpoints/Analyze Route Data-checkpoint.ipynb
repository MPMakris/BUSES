```python
%matplotlib inline
import numpy as np
import pandas as pd
import os
from geopy.distance import vincenty
import time
```

```python
agency_number = 1
route_number = 3
if type(agency_number) == int:
    agency_number = str(agency_number)
if type(route_number) == int:
    route_number = str(route_number)
```

```python
EXPECTS_ROUTE = pd.read_csv('../Data/EXPECTS_Databases_Routes/' + agency_number + '/' + 'EXPECTS_' + agency_number + '_Route_' + route_number + '.txt', sep=',')
INSTANCES_ROUTE = pd.read_csv('./INSTANCES_Databases_Routes/' + agency_number + '/' + 'INSTANCES_' + agency_number + '_Route_' + route_number + '.txt', sep=',')
STOPS = pd.read_csv('./Agency-Info/' + agency_number + '_gtfs/stops.txt', sep=',')
```

```python
Trip_Id = '1_30940700'
Stop_Id = '1_12890'
print Trip_Id, ",", Stop_Id
```

```python
INSTANCES_TRIP = INSTANCES_ROUTE[INSTANCES_ROUTE['Trip_Id']==Trip_Id]
```

```python
Stop_Loc = EXPECTS_ROUTE[EXPECTS_ROUTE['Stop_Id']==Stop_Id][['Stop_Lat', 'Stop_Lon']].iloc[0].tolist()
```

```python
Distances = []
for i in range(len(INSTANCES_TRIP)):
    Vehicle_Loc = INSTANCES_TRIP[['Vehicle_Latitude','Vehicle_Longitude']].iloc[i].tolist()
    Distances.append(vincenty(Vehicle_Loc, Stop_Loc).feet)
```

```python
INSTANCES_TRIP = INSTANCES_TRIP.assign(Distance_to_Stop=Distances)
sum(INSTANCES_TRIP['Distance_to_Stop']<=100)
```

### Analyze Just One Stop Manually

```python
%matplotlib inline
import numpy as np
import pandas as pd
from geopy.distance import vincenty
```

```python
agency_number = 1
if type(agency_number) == int:
    agency_number = str(agency_number)
route_number = 2
if type(route_number) == int:
    route_number = str(route_number)
stop_number = 450
#trip_number = 
if type(stop_number) == str:
    stop_number = int(stop_number)
Route_to_Analyze = agency_number + '_' + route_number
#Trip_to_Analyze = 
Instances_By_Route_File_Loc = '../Data/INSTANCES_Databases/Route_By_Agency/Agency_' + agency_number + '/INSTANCES_Agency_'+ agency_number +'_Route_' + route_number + '.txt'

Agency_Info_Loc = '../Agency-Info/'
Stops_File_Loc = Agency_Info_Loc + agency_number + '_gtfs/stops.txt'
```

```python
INSTANCES_To_Analyze = pd.read_csv(Instances_By_Route_File_Loc)
Stops_Info = pd.read_csv(Stops_File_Loc)
INSTANCES_To_Analyze.head()
```

```python
Stop = Stops_Info[Stops_Info['stop_id'] == stop_number]
Stop_Lat = Stop.stop_lat.values[0]
Stop_Lon = Stop.stop_lon.values[0]
Stop_Loc = (Stop_Lat,Stop_Lon)
```

```python
INSTANCES_To_Analyze.Trip_Id.unique()
```

```python
list1 = zip(INSTANCES_To_Analyze.Vehicle_Latitude.values,INSTANCES_To_Analyze.Vehicle_Longitude.values)
```

```python
Distances = []
for i in list1:
    Distances.append(vincenty(i, Stop_Loc).feet)
```

```python
INSTANCES_To_Analyze['Distance_To_Stop_'+str(stop_number)] = Distances
```
